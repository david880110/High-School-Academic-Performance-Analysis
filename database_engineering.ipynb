```python
# Ddd dependencies
import pandas as pd

from sqlalchemy import create_engine
from sqlalchemy.orm import Session
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

from sqlalchemy import Column, Integer, String, Float, Date
```

```python
# Load csv files
schools_data = 'raw_data/schools_complete.csv'
students_data = 'raw_data/students_complete.csv'
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(schools_data)
student_df = pd.read_csv(students_data)
```

```python
school_df.head()
```

```python
school_df.rename(columns={'School ID': 'school_id'}, inplace=True)
school_df.head()
```

```python
student_df.head()
```

```python
student_df.rename(columns={'Student ID': 'student_id'}, inplace=True)
student_df.head()
```

```python
# Create the Schools_data class
class Schools_table(Base):
    __tablename__ = 'schools_table'
    index = Column(Integer, primary_key = True)
    school_id = Column(Integer)
    school_name = Column(String(255))
    type = Column(String(255))
    size = Column(Float)
    budget = Column(Float)
    
# Create the Students_table class
class Students_table(Base):
    __tablename__ = 'students_table'
    index = Column(Integer, primary_key = True)
    student_id = Column(Integer)
    student_name = Column(String(255))
    gender = Column(String(255))
    grade = Column(String(255))
    school_name = Column(String(255))
    reading_score = Column(Float)
    math_score = Column(Float)
```

```python
# Create the datbase and add the classes
engine = create_engine("sqlite:///database.sqlite", echo = False)
conn = engine.connect()
session = Session(engine)

Base.metadata.create_all(conn)
```

```python
# Send the station dataframe to the sqlite database
school_df.to_sql('schools_table', engine, if_exists='replace')
student_df.to_sql('students_table', engine, if_exists='replace')
```
