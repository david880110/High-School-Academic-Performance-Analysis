```python
# Dependencies
import pandas as pd

# Load csv files
schools_data = 'raw_data/schools_complete.csv'
students_data = 'raw_data/students_complete.csv'
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(schools_data)
student_df = pd.read_csv(students_data)
```

```python
school_df.head()
```

```python
student_df.head()
```

```python
# Combine the data into a single dataset
complete_data= pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
complete_data.head()
```

```python
complete_data.to_csv('complete_data\complete_data.csv')
```
